```python
SYSTEM_PROMPT = """You're a project manager's working for a world-class startup accelerator, your name is Bob. You're managing a cohort of startups and mentors in an incubator and needs to get a good sense what each startup needs help with and what the mentors are best suited to help with. You're going to use this information to pair them. To gather the info, you're doing interviews that are transcribed. Each interview is about 20min and ends up being about 3000 words.

What you need to do next is generate a transcript as if you were Bob interviewing each of the startups/mentors you outlined above that you will write to a canvas.

Make sure that the interview is at least 20min long, it is very important."""
```

```python
MENTOR_PROMPT = """
    Here's an example of how the transcript should look like:
    ---
    [00:00:00] Bob: Hello Claire, thanks for speaking with me today. I’d like to learn about your background and how you mentor startups.
    [00:00:05] Claire: Happy to be here, Bob. I’m Claire Dupont, ex-COO of a YC-backed agri-tech startup, now independent coach based in Nairobi.
    [00:00:14] Bob: Great. Can you share your experience at the agri-tech startup?
    [00:00:17] Claire: I joined as Head of Operations in 2018. We scaled from 5 to 40 employees across three African markets, built partnerships with local cooperatives, and grew ARR from $200k to $3M in two years.
    [00:00:34] Bob: Impressive. What operational challenges did you tackle first?
    [00:00:38] Claire: Establishing reliable data collection networks was key. We had to train field agents, set up LoRaWAN gateways, and ensure data integrity despite connectivity issues in remote areas.
    [00:00:52] Bob: How did you optimize field workflows?
    [00:00:55] Claire: We mapped customer journeys, created simplified mobile apps in English and Swahili, and ran in-person training. We iterated based on feedback, reducing data entry errors by 40%.
    [00:01:14] Bob: What processes did you implement for scaling operations?
    ...
    ---
    
    *Make sure to generate a full transcript of a 20min interview, with at least 3000 words.*
    
    This is the mentor you will be interviewing:
"""
```

```python
import os

def generate_mentor_transcript(mentor):
    mentor_name = mentor.split("\n")[0].lower().replace(" ", "-").replace(",", "").replace(".", "")
    mentor_transcript = f"./test/transcripts/mentors/{mentor_name}.txt"
    if os.path.exists(mentor_transcript):
        print(f"Transcript for {mentor_name}.txt already exists, skipping generation.")
        return
    print("Generating transcript for:", mentor_name)
    response = mistral.chat.complete(
        model="mistral-large-latest",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": f"{MENTOR_PROMPT}\n\n{mentor}"},
        ],
        max_tokens=200000,
        temperature=0.5,
    )
    print("Transcript generated for:", mentor_name, "writing to file...")
    with open(mentor_transcript, "w") as f:
        f.write(response.choices[0].message.content)
    print(f"Transcript for {mentor_name}.txt saved.")
```

```python
generate_mentor_transcript(mentors[0])
```

```python
from concurrent.futures import ThreadPoolExecutor


with ThreadPoolExecutor(max_workers=3) as executor:
    futures = [executor.submit(generate_mentor_transcript, mentor) for mentor in mentors]
    for future in futures:
        try:
            future.result()  # Wait for the thread to complete
        except Exception as e:
            print(f"An error occurred: {e}")
        else:
            print("Transcript generated successfully.")
    
```
